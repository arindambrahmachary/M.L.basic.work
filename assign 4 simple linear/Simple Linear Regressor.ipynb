```python
import numpy as np
import pandas as pd
df = pd.read_csv("delivery_time.csv")
df
df.info()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5, 3))
df.plot(kind='scatter', x='Sorting Time', y='Delivery Time')
```

```python
import seaborn as sns
plt.figure(figsize=(5, 3))
sns.boxplot(x='Sorting Time', data=df)
plt.title('Boxplot of X for each Y value')
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(5, 3))
sns.boxplot(y='Delivery Time', data=df)
plt.title('Boxplot of X for each Y value')
plt.show()
```

```python
plt.figure(figsize=(4, 2))
sns.histplot(x='Sorting Time', data=df, bins=30, kde=True)
plt.title('Histogram of X for each Y value')
plt.show()
```

```python
plt.figure(figsize=(4, 2))
sns.histplot(x='Delivery Time', data=df, bins=30, kde=True)
plt.title('Histogram of X for each Y value')
plt.show()
```

```python
df['Sorting Time'].skew()
```

```python
df['Sorting Time'].kurt()
```

```python
X=df['Sorting Time']
```

```python
Y=df['Delivery Time']
```

```python
X_1=np.array(X).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(X_1,Y)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
Y_predict= lr.predict(X_1)
```

```python
Y_predict
```

```python
plt.figure(figsize=(5, 3))
plt.scatter(X_1,Y)
plt.scatter(X_1,Y_predict)
plt.plot(X_1,Y_predict,color='black')
plt.xlabel("sorting time")
plt.ylabel("delivery_time")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y,Y_predict)
print("Mean squared Error", mse.round(3))

import numpy as np
print("Root Mean squared Error", np.sqrt(mse).round(3))
```

```python
#2) Salary_hike -> Build a prediction model for Salary_hike
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("Salary_Data.csv")
df
df.info()
```

```python
X=df['YearsExperience']
Y=df['Salary']
```

```python
X_1=np.array(X).reshape(-1,1)
```

```python
plt.figure(figsize=(4, 2))
plt.scatter(X_1,Y)
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.show()

```

```python
plt.figure(figsize=(5,3))
sns.boxplot(x='Salary', data=df)
plt.title('Boxplot of X for each Y value')
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.histplot(x=X, data=df, bins=30, kde=True)
plt.title('Histogram of X for each Y value')
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.histplot(x=Y, data=df, bins=30, kde=True)
plt.title('Histogram of X for each Y value')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(X_1,Y)

Y_predict= lr.predict(X_1)
Y_predict
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
plt.figure(figsize=(5, 3))
plt.scatter(X_1,Y)
plt.scatter(X_1,Y_predict)
plt.plot(X_1,Y_predict,color='black')
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y,Y_predict)
print("Mean squared Error", mse.round(3))

import numpy as np
print("Root Mean squared Error", np.sqrt(mse).round(3))
```
